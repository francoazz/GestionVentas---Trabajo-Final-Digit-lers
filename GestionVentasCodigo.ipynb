Importa numpy y pandas

```python
import numpy as np
import pandas as pd
```

Importa el data set, indicando el delimitador correspondiente

```python
df_ventas_totales = pd.read_csv("../DataSets/VENTAS_TOTALES.csv", delimiter=";")
```

Imprime variable para corroborar que el archivo se importó correctamente

```python
df_ventas_totales
```

Imprime la información del DataFrame para ver el tipo de dato de las columnas

```python
df_ventas_totales.info()
```

Normaliza los campos descriptivos y descarta los valores NaN

```python
df_ventas_totales.dropna(inplace = True)
df_ventas_totales["SEGMENTO"] = df_ventas_totales["SEGMENTO"].str.upper()
df_ventas_totales["TIPOVENTA"] = df_ventas_totales["TIPOVENTA"].str.upper()
df_ventas_totales["NOMBREPRODUCTO"] = df_ventas_totales["NOMBREPRODUCTO"].str.upper()
```

Imprime el DataFrame para corroborar que los cambios se aplicaron correctamente

```python
df_ventas_totales
```

Recalcula la columa IMPORTEVENTAS

```python
df_ventas_totales["IMPORTEVENTAS"] = df_ventas_totales["CANTIDADUNIDADES"] * df_ventas_totales["PRECIOVENTA"]
```

Elimina la columna FECHAVENTA

```python
df_ventas_totales.drop(["FECHAVENTA"], axis=1, inplace = True)
```

Crea la variable df_productos y vuelca los datos recuperados del groupby

```python
df_productos = df_ventas_totales[["PRODUCTO", "NOMBREPRODUCTO", "IMPORTEVENTAS"]].groupby(["NOMBREPRODUCTO","PRODUCTO"]).agg("sum")
```

Imprime la nueva variable para corroborar que la operación se realizó correctamente

```python
df_productos 
```

Exporta la variable df_productos a .csv

```python
df_productos.to_csv('PRODUCTOS_FRANCOAZZETTI.csv', sep=';')
```

Crea la variable df_ventas y le carga los datos correspondientes

```python
df_ventas = df_ventas_totales[["SEGMENTO", "TIPOVENTA", "PRODUCTO", "CANTIDADUNIDADES", "PRECIOVENTA", "IMPORTEVENTAS", "DIAVENTA", "MESVENTA", "ANIOVENTA"]]
```

Imprime la variable para corroborar la carga correcta de los datos

```python
df_ventas
```

Exporta la variable df_ventas a .csv

```python
df_ventas.to_csv('VENTAS_FRANCOAZZETTI.csv', sep=';')
```
